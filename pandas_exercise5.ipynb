## Trick 38 : Pandas datetime (lot's of examples)

```
df = generate_sample_data_datetime().reset_index()
df = df.sample(500)

df
```

```
df["Year"] = df["index"].dt.year
df["Month"] = df["index"].dt.month
df["Day"] = df["index"].dt.day
df["Hour"] = df["index"].dt.hour
df["Minute"] = df["index"].dt.minute
df["Second"] = df["index"].dt.second
df["Nanosecond"] = df["index"].dt.nanosecond


df
```

```
df["Date"] = df["index"].dt.date               # 날짜만
df["Time"] = df["index"].dt.time               # 시간만
df["Time_Time_Zone"] = df["index"].dt.timetz   # 시간과 시간대를 포함
df["Day_Of_Year"] = df["index"].dt.dayofyear   # 그 해의 몇 번째 날
df["Week_Of_Year"] = df["index"].dt.weekofyear # 그 해의 몇 번째 주
df["Week"] = df["index"].dt.week  
df["Day_Of_week"] = df["index"].dt.dayofweek # 0~6 : monday~sunday
df["Week_Day"] = df["index"].dt.weekday # same as above one

df
```

```
df["Week_Day_Name"] = df["index"].dt.day_name()          # 요일이름(문자)
df["Quarter"] = df["index"].dt.quarter                   # 분기(숫자)
df["Days_In_Month"] = df["index"].dt.days_in_month       # 한달에 며칠까지 있는지
df["Is_Month_Start"] = df["index"].dt.is_month_start     # 월 시작날인지 여부
df["Is_Month_End"] = df["index"].dt.is_month_end         # 월 끝날인지 여부
df["Is_Quarter_Start"] = df["index"].dt.is_quarter_start # 분기의 시작인지
df["Is_Quarter_End"] = df["index"].dt.is_quarter_end     # 분기의 끝인지
df["Is_Leap_Year"] = df["index"].dt.is_leap_year         # 윤년인지 여부

df.head()
```

## Trick 36 : Convert from UTC to another timezone  

* Coordinated Universal Time) 또는 UTC(협정 세계표준시)  
*  UTC(협정 세계표준시)는 1972년 1월 1일부터 시행된 국제 표준시이다. UTC는 국제원자시와 윤초 보정을 기반으로 표준화되었다.  
* UTC는 그리니치 평균시(GMT)에 기반하므로 GMT로도 불리기도 하는데, UTC와 GMT는 초의 소숫점 단위에서만 차이가 나기 때문에 일상에서는 혼용된다. 기술적인 표기에서는 UTC가 사용된다.

```
s = pd.Series(range(1552194000, 1552212001, 3600)) # start, end, by

s
```

```
s = pd.to_datetime(s, unit = "s") # second까지 표현 ns면 더 자세히 표현

s
```

## Trick 32 : Filter a df with query and avoid intermediate variables

```
df = generate_sample_data()[:10]

df
```

```
df["A"] = pd.Series(["APP", "GOO", "APP", "GOO", "MIC", "MIC", "APP", "GOO", "MIC", "APP"])

df
```

```
df.rename(columns = {"A":"stock"}, inplace = True) 
'''
inplace = True로 해야 df수정사항에 반영된다. 
False로 하면 할당하지 않으면 수정된 컬럼이 df에 반영되지 않는다.
'''
print("Original df")
df
```

```
print("Filter data using intermediate variables")
temp = df.groupby("stock").mean()
temp 
```

```
temp["B"].sort_values(ascending = False)
```

```
# filter by the second greates. This way every time we generate sample data we will have a result
fv = temp["B"].sort_values(ascending = False)[1] # 내림차순으로 정렬하고 2번째 것을 뽑으면 두번째로 큰 수
temp[temp["B"] < fv]
```

```
print("Filter using query")
df.groupby("stock").mean().query("B < {}".format(fv))
df.groupby("stock").mean().query("B < @fv") # local variable을 query에서 참조시 @사용
df.groupby("stock").mean().query("B < 10") # 10보다 작은 것은 없다.
```
